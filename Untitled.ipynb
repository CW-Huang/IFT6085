```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import gzip
import cPickle as pickle
```

```python

def logreader(update):
    with gzip.open("%s_method_log.pkl.gz" % update) as f:
        cost_values = []
        try:
            while True:
                yield pickle.load(f)
        except EOFError:
            pass
```

```python
plots = {}
for update in ['sgd', 'adam', 'momentum', 'sgd_svrg']:
    cost = []
    variance = []
    for i, log in enumerate(logreader(update)):
        if 'cost' in log:
            cost.append([i, log['cost']])
        if 'grad_variance' in log:
            var = log["grad_variance"]
            variance.append([i, np.sqrt(np.sum(var[0]**2)+np.sum(var[1]**2))])

    cost = np.array(cost)
    variance = np.array(variance)
    plots[update] = (cost, variance)
```

```python
plt.figure(figsize=(10, 8))
for u in ['adam', 'momentum', 'sgd', 'sgd_svrg']:
    data = plots[u][0]
    plt.plot(data[:, 0], data[:, 1], label=u)
plt.legend()
```
